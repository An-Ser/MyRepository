```python
import numpy as np

def game_core(number):
    '''Сформируем список, запишем в него верхнюю и нижнюю границу поиска,
       для первой попытки возьмем середину между ними. Затем с каждой итерацией будем сужать интервал поиска, 
       пока не определим загаданное число'''
    count = 1
    predicts = [0,1000]
    predict = int(np.mean(predicts))
    
    while number != predict: # пока не угадали, выполняем цикл
        predicts.append(predict)
        predicts.sort() # добавляем неудавшуюся попытку в список поиска и сортируем, чтобы числа шли по порядку 
        count+=1
        a = 1 if number > predict else -1
        for elem in predicts: # найдем в списке интервал, в котором находится искомое число
            if elem == predict:
                a,b = predicts.index(elem), a + predicts.index(elem) # определим соответствующие интервалу индексы списка
                break
        predict = int(0.5*(predicts[a]+predicts[b])) # новая попытку сделаем в середину интервала
    return(count) # выходим из цикла, если угадали
              
def score_game(game_core):
    '''Чтобы исключить фактор случайности при выводе об эффективности поиска проверим наш алгоритм 1000 раз'''
    
    random_array = [] # сюда запишем случайные числа от 1 до 1000
    count_ls = [] # сюда будем записывать номера удачных попыток при угадывании числа
    random_array = [np.random.randint(0,1000) for i in range(1000)] # заполняем массив случайными числами

    for number in random_array:
        count_ls.append(game_core(number)) # добавим полученное от первой функции значение в массив результатов
    score = int(np.mean(count_ls)) # определим средний результат
    print(f"Ваш алгоритм угадывает число в среднем за {score} попыток")
    return(score)

score_game(game_core)
print('Hello after initial commit!!!!!')
print('Hello after second commit!!!')
print('test')
print('test2')
print('test3')
```
